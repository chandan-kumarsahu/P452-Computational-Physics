Centre of mass problem
Given $\lambda=x^2$ and length of rod = 2 meter. Cantre of mass can be found as 
$$ \text{centre of mass = } \dfrac{\int_0^2\lambda dx}{\int_0^2 dx} = \dfrac{\int_0^2\lambda dx}{b-a}$$


```python
def f_1(x):
    return x**2

eps=1e-5
a=0
b=2

#######################################################################################################
#    Chandan Kumar Sahu
#    1911055
#######################################################################################################

COM = int_simpson(f_1, a, b, eps) / (b-a)
print("Centre of mass of the rod is = ", COM)
```
